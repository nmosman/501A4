```
testHeartDir = "test_heart.csv"
trainingHeartDir = "training_heart.csv"

dfTest = pd.read_csv(testHeartDir)
dfTrain = pd.read_csv(trainingHeartDir)
dfTest.head()
```

```
dfTrain['famhist'] = pd.Categorical(dfTrain['famhist'])
dfTrain['famhist'] = dfTrain.famhist.cat.codes
dfTrain.pop('row.names')
chdTrain = dfTrain.pop('chd')

dfTest['famhist'] = pd.Categorical(dfTest['famhist'])
dfTest['famhist'] = dfTest.famhist.cat.codes
dfTest.pop('row.names')
chdTest = dfTest.pop('chd')
dfTrain.head()

```

```
dfTrain.dtypes
```

```
dfTest.dtypes
```

```
train_dataset = tf.data.Dataset.from_tensor_slices((dfTrain.values, chdTrain.values))
test_dataset = tf.data.Dataset.from_tensor_slices((dfTest.values, chdTest.values))
```

```
tf.constant(dfTest['famhist'])
tf.constant(dfTrain['famhist'])
train_dataset = train_dataset.shuffle(len(dfTrain)).batch(1)
test_dataset = test_dataset.shuffle(len(dfTest)).batch(1)
```

```



print("--Make model--")
model = tf.keras.models.Sequential([

  tf.keras.layers.Dense(16, activation='relu'),
  tf.keras.layers.Dense(16, activation='relu'),
  tf.keras.layers.Dense(1, activation='sigmoid')
])

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

print("--Fit model--")
model.fit(train_dataset, epochs=1000, verbose=2)

print("--Evaluate model--")
model_loss, model_acc = model.evaluate(test_dataset, verbose=2)
print(f"Model Loss:    {model_loss:.2f}")
print(f"Model Accuray: {model_acc*100:.1f}%")
```
